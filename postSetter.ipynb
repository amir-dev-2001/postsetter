```python
solder_list = []
name_list = ["ali","reza", "mamad",'dian', 'yosef', 'akbar', "aa", "bb"]
one_time = ["aa", "bb"]

while len(solder_list) < 20:
    for solder in name_list:
        if solder in one_time:
            if solder in solder_list:
                continue
            else:
                solder_list.append(solder)
        else:
            solder_list.append(solder)


chunked_list = list()
chunk_size = 9
for i in range(0, len(solder_list), chunk_size):
    chunked_list.append(solder_list[i:i+chunk_size])
print(chunked_list)
```
